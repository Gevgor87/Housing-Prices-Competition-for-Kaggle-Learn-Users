```python
y = data_all["SalePrice"][:1460]
```

```python
data_all = data_all.drop("SalePrice", axis=1)
```

```python
data_all = pd.get_dummies(data_all, drop_first=True)
```

```python
df_train_X = data_all[0:1460]
```

```python
df_test = data_all[1460:]
```

```python
df_test
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
standartsc = StandardScaler()
# X_train, X_val, y_train, y_val = train_test_split(df_train_X, y, test_size=0.2, random_state=42)
```

```python
X_train = standartsc.fit_transform(df_train_X)
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num=11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]  # Create the random grid

random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf_model = RandomForestRegressor()

# Random search of parameters, using 5 fold cross validation,
# search across 150 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator=rf_model, param_distributions=random_grid, n_iter=150,
                               cv=5, verbose=2, random_state=42, n_jobs=-1)  # Fit the random search model
```

```python
X_train
```

```python
rf_random.fit(X_train, y)
```

```python
X_test = standartsc.transform(df_test)
```

```python
rf_random.best_estimator_
```

```python
rf_random.best_params_
```

```python
rf_val_predictions = rf_random.best_estimator_.predict(X_test)
```

```python
res = pd.DataFrame({'Id': data_test.Id, 
                    'SalePrice': rf_val_predictions})
res.to_csv("submission_2.csv",index=False)
```
